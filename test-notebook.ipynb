# Code

Below are a series of code blocks that produce visualizations of the equation $z = \cos(x) \sin(y)$.

```python
from math import pi

import numpy as np
import pandas as pd
from matplotlib import pyplot as plt

%matplotlib inline
```

```python
def f(x,y):
    # multiply by pi/180 to convert degrees to radians
    return np.cos(x*pi/180) * np.sin(y*pi/180)

# note we set the upper boundary as 361
# so that 360 get's included
x,y = np.ogrid[0:361:10, 0:361:10]
z = f(x,y)

# ravel insures the x and y are 1d arrays
# try with 'contour' rather than 'contourf'
p = plt.contourf(np.ravel(x), np.ravel(y), z)

lx = plt.xlabel("x (degrees)")
tx = plt.xticks(np.arange(0,361,45))
ly = plt.ylabel("y (degrees)")
tx = plt.yticks(np.arange(0,361,45))
None
```

```python
from mpl_toolkits.mplot3d import Axes3D
fig = plt.figure()
ax = Axes3D(fig)

# create x,y grid
x,y = np.meshgrid(np.arange(0,361,10), np.arange(0,361,10))
z = f(x,y) # uses fxn f from previous cell
ax.plot_surface(x,y,z,rstride=2,cstride=2,cmap="jet")

# setup axes labels
ax.set_xlabel("x (degrees)")
ax.set_ylabel("y (degrees)")
ax.set_zlabel("z")
None
```
